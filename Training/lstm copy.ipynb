```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
import tensorflow as tf
import random

# Set seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)
random.seed(42)

base_dir = r'E:\Khóa luận\Data\Phase 2'
concentration_files = {
    '250500': 'preprocessed_250500.csv',
    '500500': 'preprocessed_500500.csv',
    '750500': 'preprocessed_750500.csv'
}

target_col = 'CH4RAWppm'
exog_cols = ['scdTEMP', 'scdHUMID', 'scdCO2', 'ccsTVOC']  # adjust as needed
window_size = 24
future_steps = 24 * 30  # 1 month (720 for hourly data)
prior_points = 500      # Number of real data points to show before forecast

for label, fname in concentration_files.items():
    print(f'\n========== Đang huấn luyện cho nồng độ: {label} ==========')    
    file_path = os.path.join(base_dir, fname)
    df = pd.read_csv(file_path)
    df['Date time'] = pd.to_datetime(df['Date time'])
    df = df.sort_values('Date time').reset_index(drop=True)
    for col in exog_cols:
        if col not in df.columns:
            df[col] = 0.0

    # Add calendar features
    df['hour'] = df['Date time'].dt.hour
    df['dayofweek'] = df['Date time'].dt.dayofweek
    feature_cols = exog_cols + ['hour', 'dayofweek', target_col]
    all_features = df[feature_cols].values

    # Normalize on all data
    scaler = MinMaxScaler()
    scaler.fit(all_features)
    all_scaled = scaler.transform(all_features)

    # Sequence creation for direct multi-step (y is a sequence)
    def create_multi_output_sequences(data, window=24, horizon=720):
        X, y = [], []
        for i in range(len(data) - window - horizon):
            X.append(data[i:i+window])
            y.append(data[i+window:i+window+horizon, -1])  # target is last col
        return np.array(X), np.array(y)

    # Use as much data as possible for training
    max_horizon = min(future_steps, len(all_scaled) - window_size - 1)
    X, y = create_multi_output_sequences(all_scaled, window=window_size, horizon=max_horizon)

    # Model
    model = Sequential()
    model.add(LSTM(128, input_shape=(window_size, len(feature_cols)), return_sequences=True))
    model.add(LSTM(64, return_sequences=False))
    model.add(Dropout(0.1))
    model.add(Dense(max_horizon))
    model.compile(optimizer='adam', loss='mse')

    early_stop = EarlyStopping(monitor='loss', patience=20, restore_best_weights=True)
    model.fit(
        X, y,
        epochs=300,
        batch_size=32,
        callbacks=[early_stop],
        verbose=1
    )

    # Forecast from the last available window
    last_window = all_scaled[-window_size:]
    last_window = last_window.reshape(1, window_size, len(feature_cols))
    y_future_pred_scaled = model.predict(last_window).flatten()

    # Inverse transform for plotting
    last_known_exog = last_window[0, -1, :-1]
    repeated_exog = np.tile(last_known_exog, (max_horizon, 1))
    inv_input = np.hstack([repeated_exog, y_future_pred_scaled.reshape(-1,1)])
    y_future_pred = scaler.inverse_transform(inv_input)[:, -1]

    # Time axis for plotting
    last_datetime = df['Date time'].iloc[-1]
    freq = df['Date time'].diff().mode()[0]
    future_datetimes = [last_datetime + (i+1)*freq for i in range(max_horizon)]

    # Plot last N real values and then forecast
    prior_datetimes = df['Date time'].iloc[-prior_points:]
    prior_values = df[target_col].iloc[-prior_points:]

    plt.figure(figsize=(16,7))
    plt.plot(prior_datetimes, prior_values, label='Giá trị thực tế (cuối)', color='blue')
    plt.plot(future_datetimes, y_future_pred, label='Dự báo LSTM (multi-step)', color='red', linestyle='--')
    plt.title(f'Dự báo {target_col} với LSTM (calendar+multi-step, future only) - {label}')
    plt.xlabel('Thời gian')
    plt.ylabel(target_col)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
