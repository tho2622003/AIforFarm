```python
import pandas as pd
import numpy as np
import pmdarima as pm
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import os
import warnings
import time  # Thêm dòng này

# Khai báo thư mục và các file dữ liệu
base_dir = r'E:\Khóa luận\Data\Phase 2'
preprocessed_files = {
    '250500': 'preprocessed_250500.csv',
    '500500': 'preprocessed_500500.csv',
    '750500': 'preprocessed_750500.csv'
}

warnings.filterwarnings("ignore")
target_col = 'scdCO2'
metrics_autoarima = []
metrics_sarima = []

for label, fname in preprocessed_files.items():
    print(f'\n========== Auto ARIMA & SARIMA cho nồng độ: {label} (dữ liệu preprocess) ==========') 
    file_path = os.path.join(base_dir, fname)
    data = pd.read_csv(file_path, encoding="utf-8")
    data = data.rename(columns=lambda x: x.strip())
    if 'Date time' in data.columns:
        data['Date time'] = pd.to_datetime(data['Date time'])
        data = data.sort_values('Date time')
        data = data.set_index('Date time')

    # Loại bỏ các hàng thiếu giá trị mục tiêu
    data = data.dropna(subset=[target_col])

    # Tách dữ liệu thành tập huấn luyện và kiểm tra (80/20)
    train_size = int(len(data) * 0.8)
    y_train = data[target_col].iloc[:train_size]
    y_test = data[target_col].iloc[train_size:]

    # ========== AUTO ARIMA ==========
    print("\n--- Đang huấn luyện mô hình Auto ARIMA (không mùa vụ)...")
    start_time = time.time()
    autoarima_model = pm.auto_arima(y_train, seasonal=False, stepwise=True, trace=True, suppress_warnings=True)
    train_time_autoarima = time.time() - start_time
    y_pred_autoarima = autoarima_model.predict(n_periods=len(y_test))

    mae_autoarima = mean_absolute_error(y_test, y_pred_autoarima)
    rmse_autoarima = np.sqrt(mean_squared_error(y_test, y_pred_autoarima))
    r2_autoarima = r2_score(y_test, y_pred_autoarima)
    mape_autoarima = np.mean(np.abs((y_test - y_pred_autoarima) / (np.abs(y_test) + 1e-10))) * 100

    print(f"Auto ARIMA - MAE: {mae_autoarima:.3f}")
    print(f"Auto ARIMA - RMSE: {rmse_autoarima:.3f}")
    print(f"Auto ARIMA - R2: {r2_autoarima:.3f}")
    print(f"Auto ARIMA - MAPE: {mape_autoarima:.2f}%")
    print(f"Auto ARIMA - Training time: {train_time_autoarima:.2f} giây")

    metrics_autoarima.append({
        'label': label,
        'MAE': mae_autoarima,
        'RMSE': rmse_autoarima,
        'R2': r2_autoarima,
        'MAPE': mape_autoarima,
        'TrainTime(s)': train_time_autoarima
    })

    plt.figure(figsize=(14, 6))
    plt.plot(y_test.index[:1000], y_test.values[:1000], label='Thực tế', linewidth=1)
    plt.plot(y_test.index[:1000], y_pred_autoarima[:1000], label='Dự báo Auto ARIMA', linewidth=1)
    plt.title(f'Dự báo Auto ARIMA so với Thực tế - {target_col} ({label})')
    plt.xlabel('Ngày')
    plt.ylabel(target_col)
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

    # ========== SARIMA ==========
    print("\n--- Đang huấn luyện mô hình SARIMA (có mùa vụ, D=1, m=24)...")
    sarima_order = (1, 0, 1)
    seasonal_order = (1, 1, 1, 24)  # m=24 cho dữ liệu theo giờ, nếu dữ liệu khác thì điều chỉnh lại

    start_time = time.time()
    sarima_model = SARIMAX(y_train, order=sarima_order, seasonal_order=seasonal_order)
    sarima_fit = sarima_model.fit(disp=False)
    train_time_sarima = time.time() - start_time
    y_pred_sarima = sarima_fit.forecast(steps=len(y_test))

    mae_sarima = mean_absolute_error(y_test, y_pred_sarima)
    rmse_sarima = np.sqrt(mean_squared_error(y_test, y_pred_sarima))
    r2_sarima = r2_score(y_test, y_pred_sarima)
    mape_sarima = np.mean(np.abs((y_test - y_pred_sarima) / (np.abs(y_test) + 1e-10))) * 100

    print(f"SARIMA - MAE: {mae_sarima:.3f}")
    print(f"SARIMA - RMSE: {rmse_sarima:.3f}")
    print(f"SARIMA - R2: {r2_sarima:.3f}")
    print(f"SARIMA - MAPE: {mape_sarima:.2f}%")
    print(f"SARIMA - Training time: {train_time_sarima:.2f} giây")

    metrics_sarima.append({
        'label': label,
        'MAE': mae_sarima,
        'RMSE': rmse_sarima,
        'R2': r2_sarima,
        'MAPE': mape_sarima,
        'TrainTime(s)': train_time_sarima
    })

    plt.figure(figsize=(14, 6))
    plt.plot(y_test.index[:1000], y_test.values[:1000], label='Thực tế', linewidth=1)
    plt.plot(y_test.index[:1000], y_pred_sarima[:1000], label='Dự báo SARIMA', linewidth=1)
    plt.title(f'Dự báo SARIMA so với Thực tế - {target_col} ({label})')
    plt.xlabel('Ngày')
    plt.ylabel(target_col)
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

# Tổng hợp kết quả các mô hình Auto ARIMA
metrics_df_autoarima = pd.DataFrame(metrics_autoarima)
print("\nTổng hợp kết quả các mô hình Auto ARIMA:")
print(metrics_df_autoarima)

# Tổng hợp kết quả các mô hình SARIMA
metrics_df_sarima = pd.DataFrame(metrics_sarima)
print("\nTổng hợp kết quả các mô hình SARIMA:")
print(metrics_df_sarima)
```
